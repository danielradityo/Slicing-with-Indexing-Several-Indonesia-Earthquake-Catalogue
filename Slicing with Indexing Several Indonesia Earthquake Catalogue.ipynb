```python
# Import Libraty Pandas
import pandas as pd
```

```python
# Read Dataset
data = pd.read_excel('Indonesia Eartquake Catalogue Edit.xlsx')
data
```

```python
# Get Unique Value
data.a.value_counts().head(50)
```

```python
# Slicing Several data (dataleft and dataright)
dataleft = data[(data['a'] == 'ISC') | (data['a']== 'MLE') | (data['a']== 'MLI') | (data['a']== 'PDE') | (data['a']== 'PDEQ') | (data['a']== 'REB') | (data['a']== 'SWEQ') | (data['a']== 'PDEW')]
dataright = data[(data['a']=='V10')]
```

```python
dataleft
```

```python
# Get several data with indexing
dataleft.iloc[:, 8:17]
```

```python
# Drop un-use columns
dataleft=dataleft.drop(columns=dataleft.columns[8:17])
```

```python
dataleft
```

```python
# Get several data with indexing
dataright.iloc[:, 0:11]
```

```python
# Drop un-use columns
dataright=dataright.drop(columns=dataright.columns[0:11])
```

```python
dataright
```

```python
# Export to excel
dataleft.to_excel('dataleft.xlsx')
dataright.to_excel('dataright.xlsx')
```
